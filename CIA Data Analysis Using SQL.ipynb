```python
%%sql
SELECT *
  FROM sqlite_master
 WHERE type='table';
```

```python
%%sql
SELECT *
  FROM facts
limit 5;
```

```python
%%sql
SELECT Min(population), MAX(population), MIN(population_growth), MAX(population_growth)
  FROM facts;
```

```python
%%sql
SELECT name, MIN(population)
  FROM facts
 GROUP BY name
ORDER by population asc;
```

```python
%%sql
SELECT name, MAX(population)
  FROM facts
 GROUP BY name
ORDER by population desc;
```

```python
%%sql
SELECT Min(population), MAX(population), MIN(population_growth), MAX(population_growth)
    FROM facts
    Where Population < 7256490011 AND Population > 1;
```

```python
%%sql
SELECT AVG(population)
    FROM facts
    Where Population < 7256490011 AND Population > 1;
```

```python
%%sql
SELECT AVG(area)
    FROM facts
    Where Population < 7256490011 AND Population > 1;
```

```python
%%sql
SELECT name, population AS avg_population
  FROM facts
    WHERE Population > (SELECT avg(population)
                         FROM facts
                       Group BY Name);
```

```python
%%sql
SELECT name, area AS avg_area
  FROM facts
    WHERE area > (SELECT avg(area)
                         FROM facts
                       Group BY Name);
```

```python
%%sql
SELECT name, CAST(area_land AS FLOAT)/CAST(area_water AS FLOAT) AS land_water_ratio
  FROM facts
    ORDER BY land_water_ratio desc;
```

```python
%%sql
SELECT name, population_growth
  FROM facts
    ORDER BY population_growth desc
    limit 5;
```

```python
%%sql
SELECT name, birth_rate 
  FROM facts
    WHERE birth_rate - death_rate > 0
    ORDER BY birth_rate asc;
```
